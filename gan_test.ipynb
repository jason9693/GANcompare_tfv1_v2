```python
import tensorflow as tf
import numpy as np
from tensorflow.examples.tutorials.mnist import input_data
from Network.GAN import *
from matplotlib import pyplot as plt
import matplotlib.gridspec as gridspec

%matplotlib inline
plt.rcParams['figure.figsize'] = (10.0, 8.0) # set default size of plots
plt.rcParams['image.interpolation'] = 'nearest'
plt.rcParams['image.cmap'] = 'gray'

def show_images(images):
    images = np.reshape(images, [images.shape[0], -1])  # images reshape to (batch_size, D)
    sqrtn = int(np.ceil(np.sqrt(images.shape[0])))
    sqrtimg = int(np.ceil(np.sqrt(images.shape[1])))

    fig = plt.figure(figsize=(sqrtn, sqrtn))
    gs = gridspec.GridSpec(sqrtn, sqrtn)
    gs.update(wspace=0.05, hspace=0.05)

    for i, img in enumerate(images):
        ax = plt.subplot(gs[i])
        plt.axis('off')
        ax.set_xticklabels([])
        ax.set_yticklabels([])
        ax.set_aspect('equal')
        plt.imshow(img.reshape([sqrtimg,sqrtimg]))
    return
```

```python
mnist = input_data.read_data_sets("MNIST_data/", one_hot=True)
```

```python
batch_size = 64
learning_rate = 1e-3
epoch = 100
# label_size = 10
z_dim = 96
```

```python
if True:
    
    sess = tf.Session()
    gan = GANv2(
        input_shape=list(mnist.train.images.shape[1:]),
        learning_rate=learning_rate,
        sess= sess,
        noise_dim=96,
        ckpt_path=None,
        num_classes = 1
    )
    
#     GAN(
#         input_shape=list(mnist.train.images.shape[1:]),
#         learning_rate=learning_rate,
#         sess= sess,
#         noise_dim=96,
#         ckpt_path=None,
#         num_classes = 1
#     )
    
    g_loss_history = []
    d_loss_history = []
   
    for i in range(epoch):
      g_loss_mean = 0
      d_loss_mean = 0
      
      for j in range(mnist.train.num_examples // batch_size):
        images,labels = mnist.train.next_batch(batch_size)
        labeles = np.argmax(labels,axis = 1)
        # images = mnist.train.images[:batch_size]
        # labeles = mnist.train.labeles[:batch_size] 
        random_uniform = np.random.uniform(-1,1,size=[batch_size,z_dim])
        d_loss = gan.train(x=images, z=random_uniform)
        #op,loss,kld,likelihood
        d_loss_mean+= d_loss
        
        #for k in range(1 + 10 * int(j==0)):
        if True:
            g_loss = gan.train(z=random_uniform)
            random_uniform = np.random.uniform(-1,1,size=[batch_size,z_dim])
        
        g_loss_mean += g_loss
        
        if (j == 1):
            # print('kld : ',kld,' likelihood: ',likelihood)
            print('--------------------------')
            print('g loss in {}epoch, {}batch : {}'.format(i,j, g_loss))
            print('d loss in {}epoch, {}batch : {}'.format(i, j ,d_loss))
            print('eval:')
            gan.eval(np.random.uniform(-1,1,[16,z_dim]))
            plt.show()
            print('--------------------------')

      g_loss_mean /= (mnist.train.num_examples // batch_size)
      d_loss_mean /= (mnist.train.num_examples // batch_size)
      
      g_loss_history.append(g_loss_mean)
      d_loss_history.append(d_loss_mean)
        
    
#     plt.plot(d_loss_history)
#     plt.plot(g_loss_history)
#     plt.legend(['d_loss','g_loss'],loc='upper left')
#     # 
    # Z = np.random.normal(size=(batch_size,label_size))
    # rand_y = np.random.randint(0,10,(batch_size))
    # predict = vae.predict(Z,rand_y)

```
